# Exploratory data analysis for all files


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Defining the DataAnalysis class
class DataAnalysis:
    def __init__(self, file_path):
        """Initialize the DataAnalysis object with the file path."""
        self.file_path = file_path
        self.data = None
    
    def load_data(self):
        """Load CSV data into a DataFrame."""
        self.data = pd.read_csv(self.file_path)
        print(f"Data loaded successfully from {self.file_path}")
    
    def basic_info(self):
        """Print basic information about the data."""
        if self.data is None:
            raise ValueError("Data not loaded. Please load the data first.")
        
        print(f"--- Basic Information for {self.file_path} ---")
        print("Shape of the data:", self.data.shape)
        print("\nData Types:\n", self.data.dtypes)
        print("\nMissing Values:\n", self.data.isnull().sum())
        print("\nBasic Statistics:\n", self.data.describe())

    def plot_distributions(self):
        """Plot distributions for numerical columns."""
        if self.data is None:
            raise ValueError("Data not loaded. Please load the data first.")
        
        numeric_cols = self.data.select_dtypes(include=['float64', 'int64']).columns
        self.data[numeric_cols].hist(bins=15, figsize=(15, 6), layout=(2, len(numeric_cols)//2))
        plt.suptitle(f"Distributions of Numerical Columns - {self.file_path}", fontsize=16)
        plt.show()

    def plot_correlations(self):
        """Plot a heatmap of the correlation matrix."""
        if self.data is None:
            raise ValueError("Data not loaded. Please load the data first.")
        
        # Select only numerical columns for correlation matrix
        numeric_data = self.data.select_dtypes(include=['float64', 'int64'])

        if numeric_data.empty:
            print(f"No numerical columns found in {self.file_path} to plot correlations.")
            return

        plt.figure(figsize=(10, 6))
        sns.heatmap(numeric_data.corr(), annot=True, fmt='.2f', cmap='coolwarm', square=True)
        plt.title(f"Correlation Matrix - {self.file_path}")
        plt.show()

    def perform_eda(self):
        """Run the full EDA process."""
        self.load_data()
        self.basic_info()
        self.plot_distributions()
        self.plot_correlations()

# List of file paths
csv_files = [
    '../data/yfinance_data/AAPL_historical_data.csv',
    '../data/yfinance_data/GOOG_historical_data.csv',
    '../data/yfinance_data/MSFT_historical_data.csv',
    '../data/yfinance_data/TSLA_historical_data.csv',
    '../data/yfinance_data/AMZN_historical_data.csv',
    '../data/yfinance_data/META_historical_data.csv',
    '../data/yfinance_data/NVDA_historical_data.csv',
]

# Performing EDA for all files in a loop
for file in csv_files:
    analysis = DataAnalysis(file)
    analysis.perform_eda()
```
